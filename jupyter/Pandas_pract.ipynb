# Merge,Join ,Inner,outer 


df1=pd.DataFrame({
     'id':[1,2,3,4],
     'name': ['keerti','Amit','Educos','yotube']
}
)

df2=pd.DataFrame({
    'id':[3,4,5,6],
     'course': ['Genai','CPP','HLD','LLD']
})

##  Inner Join (only common id)

```python
merged_inner=pd.merge(df1,df2,on='id',how='inner')
merged_inner
```

## Left Join (all df1 rows, match from df2)

```python
merged_left=pd.merge(df1,df2,on='id',how='left')
merged_left
```

## Right Join (all df2 rows, match from df1)

```python
merged_right=pd.merge(df1,df2,on='id',how='right')
merged_right
```

## Outer Join (all rows from both)

```python
merged_outer=pd.merge(df1,df2,on='id',how='outer')
merged_outer.fillna("Not Available",inplace=True)
# merged_outer.dropna()
merged_outer
```

## when id not coomon then how to use join

```python
df3=pd.DataFrame({
     'id3':[1,2,3,4],
     'name': ['keerti','Amit','Educos','yotube']
}
)

df4=pd.DataFrame({
    'id4':[3,4,5,6],
     'course': ['Genai','CPP','HLD','LLD']
})

merged_inner=pd.merge(df3,df4,left_on='id3',right_on='id4',how='inner')
merged_inner
```
